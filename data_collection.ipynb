```python
from time import sleep
from chess_ai.chess_player import RandomAgent, PieceSquaresAgent, MonteCarloAgent, MonteCarloAgentBlack
from chess_ai.chess_game import Game
from chess import svg, Board
from IPython.display import display, clear_output
import csv
from datetime import datetime
import time
```

```python
sample_size = 100
config = {
    'depth': 50,
    'path_count': 50,
    'win_weight': 1,
    'draw_weight': -1,
    'lose_weight': -2
}
save_file = f'MonteCarlo_{datetime.now()}.csv'
header = ['agent', 'winner', 'turns', 'run_time', 'depth', 'path_count', 'win_weight', 'draw_weight', 'lose_weight']
data = []

# Experiment
for i in range(sample_size):
    player_1 = MonteCarloAgent('MonteCarlo', **config)
    player_2 = RandomAgent('Random')
    g = Game()

    print(f'running game {i}')
    start_time = time.time()
    result = g.play_and_get_turns_played(player_1, player_2)
    run_time = (time.time() - start_time)

    if result[0]:
        winner = result[0].name
    else:
        winner = 'draw'
    
    data.append(['MonteCarlo', winner, result[1], run_time, config['depth'], config['path_count'], config['win_weight'], config['draw_weight'], config['lose_weight']])


with open(f'./data/{save_file}', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    # write the data
    for row in data:
        writer.writerow(row)

print('complete')
```
